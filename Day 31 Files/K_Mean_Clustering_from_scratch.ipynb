```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.random.normal(loc=[0, 0], scale=1.1, size=(100, 2))*20
x2 = np.random.normal(loc=[5, 5], scale=1, size=(100, 2))*10

X = np.concatenate((x1, x2))

centroids  = np.random.normal(loc=[0, 5], scale=1, size=(2, 2))

plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# Number of clusters
k = 2


# Number of iterations
max_iterations = 10

for iter in range(max_iterations):
    # Calculate distances from each point to each centroid
    distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))

    # Assign each point to the nearest centroid
    labels = np.argmin(distances, axis=0)

    # Update centroids
    new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])

    centroids = new_centroids
    plt.scatter(X[:, 0], X[:, 1], c=labels)
    plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='red', label='Centroids')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('K-Means Clustering:Iteration: {}'.format(iter))
    plt.legend()
    plt.show()

```
